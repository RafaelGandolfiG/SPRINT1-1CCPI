```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

Questão A

```python
#A: Gráfico da função
t = np.linspace(0, 50, 400)
p = 18 + (24 * t)/(t + 2)
#Gráfico
plt.figure()
plt.plot(t,p,label = 'P(t) = 18 + 24 * t / (t + 2)')
plt.xlabel('Tempo (min)')
plt.ylabel('Potência (kW)')
plt.title('Gráfico da função P(t)')
plt.grid()
plt.show()
```

Questão B

```python
t = sp.Symbol('t',positive = True)
P = 18 + 24 * t / (t + 2)
limite_de_0 = sp.limit(P,t,0,dir='+')
print(f'Limite quando t -> 0+: {limite_de_0}')
limite_de_infinito = sp.limit(P,t,sp.oo)
print(f'Limite quando t -> infinito: {limite_de_infinito}')

#Explique oque esses valores representam para o comportamento da recarga ao longo do tempo
#A recarga começa mais fraca (18 kW), cresce rapidamente e depois se estabiliza em 42 kW, mostrando que existe um limite físico para a potência do carregador.
```

Questão C

```python
def variacao(t):
  funcao = 18 + (24 * t)/(t + 2)
  return funcao
# Tempo == 3
tempo_3 = variacao(3)
print(f'A variação para t -> 3 é de: {tempo_3}')
# Tempo == 1
tempo_1 = variacao(1)
print(f'A variação para t -> 1 é de: {tempo_1}')
variacao_media = (tempo_3 - tempo_1) / 3 - 1
print(f'A variação média é de: {variacao_media:.2f} kW/min')
#Oque esses resultados indicam para o gerenciamento de infraestrutura?
#Isso indica que, entre 1 e 3 minutos, a potência aumenta rapidamente,exigindo que a infraestrutura elétrica suporte um crescimento significativo de carga nesse intervalo.
```
